```python
m = re.search('(\w+),(\w+),(\w+)', 'foo,quux,baz')
print(f"m.groups(): {m.groups()}")

print(f"m.group() = content: '{m.group()}'")
print("m.group(1): ", m.group(1))
print("m.group(2): ", m.group(2))
print("m.group(3): ", m.group(3))

m.group(2, 3)
m.group(3, 2, 1)
```

```python
match = re.search('(?P<w1>\w+),(?P<w2>\w+),(?P<w3>\w+)', 'foo,quux,baz')
w1 = match.group('w1')
w2 = match.group('w2')
w3 = match.group('w3')
print("match.groups(): ", match.groups())
print("match.group(): ", match.group())
print(w1, w2, w3)
```

```python
"""(?:\w+) ko ton tai trong match.groups()"""
m = re.search('(\w+),(?:\w+),(\w+)', 'foo,quux,baz')
print(m.groups())
print(m.group())
print(m[1])
print(m[2])
```
