```python
import itertools
import operator
```

```python
chars = ['a', 'b', 'c', 'd', 'e']
numbers = [1,2,3]

zip_one = itertools.zip_longest(chars, numbers)
print(list(zip_one))
```

```python
zip_two = itertools.zip_longest('ABCD', '12', fillvalue='#')
print(list(zip_two))
```

```python
# repeat items trung nhau
per_one = itertools.permutations('ABCD', 2)
print(list(per_one))
```

```python
per_two = itertools.permutations(range(3), 3)
list(per_two)
```

```python
# no repeat items trung nhau
com_one = itertools.combinations('ABCD', 3)
list(com_one)
```

```python
com_two = itertools.combinations(range(4), 3)
list(com_two)
```

```python
# ((x,y) for x in A for y in B)
pro_one = itertools.product('ABCD', repeat=2)
list(pro_one)
```

```python

chain_one = itertools.chain('ABC', '123')
list(chain_one)
```

```python
# accumulate(iterable, func=operator.add)
acc_add = itertools.accumulate([1,2,3,4,5])
list(acc_add)
```

```python
acc_mul = itertools.accumulate([1,2,3,4,5], func=operator.mul)
list(acc_mul)
```

```python
# groupby
students = [{'name': 'Paul', 'mail': '@gmail.com'}, {'name': 'Tom', 'mail': '@yahoo.com'},
            {'name': 'Jim', 'mail': '@gmail.com'}, {'name': 'Jules', 'mail': '@something.com'},
            {'name': 'Gregory', 'mail': '@gmail.com'}, {'name': 'Kathrin', 'mail': '@something.com'}]
```

```python
key_func = lambda student: student['mail']
# lay key cho groupby

students.sort(key=key_func)
# sorting by same key function we later use with groupby
students
```

```python
for key, group in itertools.groupby(students, key=key_func):
    print(f"key: {key}")
    print(f"group {list(group)}")
```

```python
for key, group in itertools.groupby([3,3,3,3,2,2,5,5,5,5,234,5,8,8,7,7,8]):
    print(f"key: {key} \ngroup: {list(group)}")
```
