# basic

```python
class NumberSequence:

    def __init__(self, start=0):
        self.current = start

    def next(self):
        current = self.current
        self.current += 1
        return current

seq = NumberSequence()
print(f"{seq.next() = }")
print(f"{seq.next() = }")
print(f"{seq.next() = }")
print(f"{seq.next() = }")
print(f"{seq.next() = }")
```

# __iter__ and __next__

```python
class SequenceOfNumbers:

    def __init__(self, start=0):
        self.current = start

    def __next__(self):
        print("__next__")
        current = self.current
        self.current += 1
        return current

    def __iter__(self):
        print("__iter__")
        return self
```

```python
seq = SequenceOfNumbers(100)
print(f"{next(seq) = }")
print(f"{next(seq) = }")
print(f"{next(seq) = }")
print(f"{next(seq) = }")
```

```python
list(zip("abcd", SequenceOfNumbers(50)))
```

```python
list(zip(SequenceOfNumbers(), "abcdef"))
```

# yield

```python
def sequence(start=0):

    while True:
        yield start
        start += 1
        
list(zip(sequence(), "abcdef"))
```

```python
list(zip("abcdef", sequence()))
```

# yield from

```python
def chain(*iterables):
    
    for iterable in iterables:
        yield from iterable
        
list(chain("hello", ["world"], ("tuple", " of ", "values.")))
```

```python
def _chain(*iterables):
    for iterable in iterables:
        
        for value in iterable:
            yield value

list(_chain("hello", ["world"], ("tuple", " of ", "values.")))
```

```python
def all_powers(n, power):
    yield from (n ** i for i in range(power + 1))
```
