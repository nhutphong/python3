```python
import time
import asyncio
import threading
import inspect
import sys

# trong jupyter phai co moi run duoc asyncio
import nest_asyncio
nest_asyncio.apply()

#-----------------------------------------------------------------------------#
#nen dung tu python3.8 syntax don gian

def child_zero():# funciton='child'
    return f"{inspect.stack()[0].function}"

def child_one():# function='name tu func goi no'
    return f"{inspect.stack()[1].function}"

def whoami_zero(): 
    return sys._getframe(0).f_code.co_name

def whoami_one(): 
    return sys._getframe(1).f_code.co_name


async def ONE(number, delay=1):
    total = 0
    for num in range(1, number+1):
        print(f"{child_one()}_ {num} START")
        await asyncio.sleep(delay)
        print(f"{child_one()}_ {num} END")
        total += num
    return total


async def TWO(chars, delay=1):
    for i, char in enumerate(chars, 1):
        print(f"{child_one()}_{i}_{char} START")
        await asyncio.sleep(delay)
        print(f"{child_one()}_{i}_{char} END")

    return chars[::-1]


async def THREE_www(number, delay=1):
    total = 0
    for num in range(1, number+1):
        print(f"{child_one()}_{num} START")
        await www('ABCD',id=num)
        await asyncio.sleep(delay)
        print(f"{child_one()}_{num} END")

        total += num

    return total

async def www(chars, id):
    for i, char in enumerate(chars, 1):
        print(f"{child_one()}_{id}_{i}_{char} START")
        await asyncio.sleep(2)
        print(f"{child_one()}_{id}_{i}_{char} END")

    return chars[::-1]


async def FOUR(chars, delay=1):
    for i, char in enumerate(chars, 1):
        print(f"{child_one()}_{i}_{char} START")
        await asyncio.sleep(delay)
        print(f"{child_one()}_{i}_{char} END")

    return chars[::-1]


async def run_async():
    print(f"{'run_async START':-^85}")
    start = time.time()
    print(f"started at {time.strftime('%X')}")

    # asyncio.gather(*coros) => *coros = so luong coros chinh
    #get_func1, get_func2 = await asyncio.gather(func1, func2, ...)
    coros = await asyncio.gather(ONE(3,1), TWO("nhu", 1),THREE_www(3, 5),FOUR('ZUT', 2))
    """
        run 1 lan ONE, TWO, THREE_www(run 1 coros con), FOUR, truoc
        sau do run cac coros co time it nhat KO QUAN TAM VI TRI
    """

    print(f"\nONE: {coros}")

    end = time.time()
    duration = end - start
    print(f"finished at {time.strftime('%X')}")
    durationf = f"{duration:,.2f}"
    print(f"{durationf:-^85}")
    print(f"\n{'run_async END':-^85}")
    return coros

asyncio.run(run_async())
```
