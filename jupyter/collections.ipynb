```python

from collections import *

# initializing dictionaries 
dic1 = { 'a' : 1, 'b' : 2 } 
dic2 = { 'b' : 3, 'c' : 4 } 
dic3 = { 'f' : 5 } 

#ChainMap() xu ly nhieu dicts
# initializing ChainMap 
chain = ChainMap(dic1, dic2) 
print(chain)
# printing chainMap using map 
print ('chain.maps: ',chain.maps)
print()

# using new_child() to add new dictionary 
chain1 = chain.new_child(dic3) 
print("chain1 = chain.new_child(dic3)")
print('chain1.maps',chain1.maps) 

# displaying value associated with b before reversing 
print ("Value associated with b before reversing is : ",end="") 
print (chain1['b']) 

# reversing the ChainMap 
chain1.maps = reversed(chain1.maps) 

# displaying value associated with b after reversing 
print ("Value associated with b after reversing is : ",end="") 
print (chain1['b']) 
```

```python
# xu ly 1 dict
d = OrderedDict.fromkeys(['thong', 'dung', 'cho', 'trau'])
print(d)
print()

d.move_to_end('thong')
print("d.move_to_end('thong') move ('thong', None) xuoi cuoi")
print(d)
print()

d.move_to_end('cho', last=False)
print("d.move_to_end('cho', last=False) move ('cho', None ) len dau")
print(d)
print()

print("d.popitem(last=False)  delete dau")
d.popitem(last=False)
print(d)
```

```python
# collections.Counter([iterable-or-mapping])¶
print("collections.Counter( [iterable|mapping|**kwds] )")
count_str = Counter("thanhdung") 

print(f"Counter('thandung'): {count_str}")
print("count_str.elements(): ",count_str.elements())
print("list(count_str.elements()): ", list(count_str.elements()))
print(end='\n'*2) 


count_kwds = Counter( a = 2, b = 3, c = 6, d = 1, e = 5) 
print("Counter( a = 2, b = 3, c = 6, d = 1, e = 5)")
print("count_kwds.elements(): ",count_kwds.elements())
print("list(count_kwds.elements()): ", list(count_kwds.elements()))



print("Counter('abracadabra').most_common(3): get 3 items co so duplicate nhieu nhat")
print(Counter('abracadabra').most_common(3))


c = Counter(a=4, b=2, c=0, d=-2)
d = Counter(a=1, b=2, c=3, d=4)
print('c: ', c)
print('d: ', d)
print('c.subtract(d) c-d')
c.subtract(d)
print(c)
print('con vai operator nhu: c + d, c & d, c | d')
```

```python

de = deque([1, 2, 3,])
print(de)
print()

de.extend([4,5,6]) 
print('de.extend([4,5,6]) ')
print(de)
print()

de.extendleft([7,8,9])
print('de.extendleft([7,8,9])')
print(de) 
print()


de.rotate(3)
print("de.rotate(3): lay 3 item CUOI quang len DAU")
print (de)
print()

de.rotate(-3)
print("de.rotate(-3) lay 3 tiem DAU quang xuong CUOI")
print (de)
print()

de.reverse() 
print('de.reverse()')
print (de) 
print()

de.appendleft("thong")
print("de.appendleft('thong')")
print(de)
print()

de.popleft()
print('de.popleft()')
print(de)
```

```python
s = 'mississippi'
d = defaultdict(int)
for k in s:
    d[k] += 1
    
print(d)
print(d.items())
print()

s = [('yellow', 1), ('blue', 2), ('yellow', 3), ('blue', 4), ('red', 1)]
d = defaultdict(list)
for k, v in s:
    d[k].append(v)

print(d)
print(d.items())
print()

    
# Defining the dict 
d = defaultdict(lambda: "Tao la lambda") 
d["a"] = 1
d["b"] = 2

# khi key khong ton tai se la: "Tao la lambda"
print(d['d']) 
print(d.__missing__('b')) 
```
